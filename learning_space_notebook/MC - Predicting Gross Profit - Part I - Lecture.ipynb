```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170 #平均
# シミュレーションに必要な指標->平均と標準偏差
#１７０の中の +- 20 に売り上げがあると考えられる
rev_stdev = 20 #標準偏差 
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev[:10]
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.show()
```

```python
#売り上げ原価
COGS = - (rev * np.random.normal(0.6,0.1)) #元の売り上げの0.6%と0.1%が標準偏差
 
plt.figure(figsize=(15, 6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```
